```python
import pandas as pd 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Select Your Stock

```python
stock_name = 'META' # ticker symbol
df = pd.read_csv('./dataset/' + stock_name + '.csv')
```

# Dataset Time period 

```python
print('\t\tStart Date\t\tEnd Date')
print('Stock\t\t{}\t\t{}'.format(df['Date'][0], df.iloc[-1, 0]))
```

# Stock's Price Plots

```python
fig,axes = plt.subplots(2,2, figsize=(15,15))
axes[0][0].plot(df['Date'], df['Close'])
axes[0][0].set_title('Close Price')
axes[0][0].set_xlabel('Time', fontsize = 10)
axes[0][1].set_ylabel('Close Price', fontsize = 10)
axes[0][0].xaxis.set_tick_params(labelsize = 8)
axes[0][0].yaxis.set_tick_params(labelsize = 8)

axes[0][1].plot(df['Date'], df['Open'])
axes[0][1].set_title('Open Price')
axes[0][1].set_xlabel('Time', fontsize = 10)
axes[0][1].set_ylabel('Open Price', fontsize = 10)
axes[0][1].xaxis.set_tick_params(labelsize = 8)
axes[0][1].yaxis.set_tick_params(labelsize = 8)


axes[1][0].plot(df['Date'], df['High'])
axes[1][0].set_title('High Price')
axes[1][0].set_xlabel('Time', fontsize = 10)
axes[1][0].set_ylabel('High Price', fontsize = 10)
axes[1][0].xaxis.set_tick_params(labelsize = 8)
axes[1][0].yaxis.set_tick_params(labelsize = 8)

axes[1][1].plot(df['Date'], df['Low'])
axes[1][1].set_title('Low Price')
axes[1][1].set_xlabel('Time', fontsize = 10)
axes[1][1].set_ylabel('Low Price', fontsize = 10)
axes[1][1].xaxis.set_tick_params(labelsize = 8)
axes[1][1].yaxis.set_tick_params(labelsize = 8)

plt.show()
```

# Moving Averages (50 and 100 days)

```python

# Calculate the 10-day moving average of the "Close" column
df['10d_ma'] = df['Close'].rolling(window=100).mean()

# Calculate the 30-day moving average of the "Close" column
df['30d_ma'] = df['Close'].rolling(window=50).mean()

# Plot the "Close" and moving average columns
plt.plot(df['Close'], label='Close')
plt.plot(df['10d_ma'], label='100-day MA')
plt.plot(df['30d_ma'], label='50-day MA')

# Add a legend and display the plot
plt.legend()
plt.show()
```

# Box Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111)

bp = ax.boxplot([df['Open'], df['High'], df['Low'], df['Close'],])

ax.set_xticklabels(['Open', 'High', 'Low', 'Close'])

plt.show()
```
